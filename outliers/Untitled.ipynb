```python
import pandas as pd 
df=pd.read_csv('height_weight.csv')
df.head()
Q1=df.weight.quantile(.25)
Q3=df.weight.quantile(.75)
IQR=Q3-Q1
```

```python
df
```

```python
print(Q1, Q3,IQR)
```

```python
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
print(lb,ub)
```

```python
df.describe()
```

```python
df_withoutoutliers=df[(df.weight>lb)&(df.weight<ub)]
df_withoutoutliers.head()
```
